# Tratamento do tipo das features


- Criação de uma feature `ESTADO` a partir do `CEP_2_DIG` com o intuito de remover as features `DDD` e `CEP_2_DIG` e separar por estados;
- `FLAG_PF` será alterado de "X" para 0 e "NaN" para 1;
- `DATA_CADASTRO`, `DATA_EMISSAO_DOCUMENTO`, `DATA_PAGAMENTO`, `DATA_VENCIMENTO` e `SAFRA_REF` estão no formato errado.

```python
df_cad_pag_info.loc[df_cad_pag_info['CEP_2_DIG'] == 'na', 'CEP_2_DIG'] = '0' # Alterando o valor na para um inteiro que não existe CEP
df_cad_pag_info['CEP_2_DIG'] = df_cad_pag_info['CEP_2_DIG'].astype(int) # Alterando o tipo da variável

display(df_cad_pag_info['CEP_2_DIG'].value_counts().sort_index())
```

```python
# Organizando os ceps por estado
cep = df_cad_pag_info['CEP_2_DIG']

condicoes = [
    (cep >= 1) & (cep <= 19),
    (cep >= 20) & (cep <= 28),
    (cep == 29),
    (cep >= 30) & (cep <= 39),
    (cep >= 40) & (cep <= 48),
    (cep == 49),
    (cep >= 50) & (cep <= 56),
    (cep == 57),
    (cep == 58),
    (cep == 59),
    (cep >= 60) & (cep <= 63),
    (cep == 64),
    (cep == 65),
    (cep >= 66) & (cep <= 68),
    (cep == 69),
    (cep >= 70) & (cep <= 72),
    (cep == 73),
    (cep >= 74) & (cep <= 76),
    (cep == 77),
    (cep == 78),
    (cep == 79),
    (cep >= 80) & (cep <= 87),
    (cep == 88) | (cep == 89),
    (cep >= 90) & (cep <= 99),
]

estados = [
    'SP', 'RJ', 'ES', 'MG', 'BA', 'SE', 'PE', 'AL', 'PB', 'RN',
    'CE', 'PI', 'MA', 'PA', 'AC', 'DF', 'DF', 'GO', 'TO', 'MT',
    'MS', 'PR', 'SC', 'RS'
]

df_cad_pag_info['ESTADO'] = np.select(condicoes, estados, default='Desconhecido')
```

```python
#@title Tratamento da feature 'FLAG_PF'
mapeamento = {'X': 0, '1': 1}
df_cad_pag_info['FLAG_PF'] = df_cad_pag_info['FLAG_PF'].map(mapeamento).fillna(1).astype(int)
```

```python
#@title Tratamento das features que representam datas
datas = ['DATA_EMISSAO_DOCUMENTO', 'DATA_PAGAMENTO', 'DATA_VENCIMENTO', 'DATA_CADASTRO']

# Todas estão no tipo object e serão transformadas para datetime, no caso de 'SAFRA_REF' será para period

df_cad_pag_info[datas] = df_cad_pag_info[datas].apply(pd.to_datetime)

df_cad_pag_info['SAFRA_REF'] = pd.to_datetime(df_cad_pag_info['SAFRA_REF']).dt.to_period('M')
```

```python
#@title Criação da target

# Seguindo o critério de 5> dias de atraso = inadimplente
df_cad_pag_info["DIAS_ATRASO"] = (df_cad_pag_info["DATA_PAGAMENTO"] - df_cad_pag_info["DATA_VENCIMENTO"]).dt.days

df_cad_pag_info["INADIMPLENTE"] = df_cad_pag_info["DIAS_ATRASO"].apply(lambda x: 1 if x >= 5 else 0)
```

```python
# Visualização das informações
display(df_cad_pag_info.info())
display(df_cad_pag_info.isnull().sum())
```

```python
#@ Criação de features para mostrar idade (em dias) da emissão e cadastro das empresas

data_ref = pd.Timestamp.now()

# Tempo desde a emissão
df_cad_pag_info['dias_desde_emissao'] = (
    data_ref - df_cad_pag_info['DATA_EMISSAO_DOCUMENTO']
).dt.days

# Idade do cadastro
df_cad_pag_info['idade_cadastro_dias'] = (
    data_ref - df_cad_pag_info['DATA_CADASTRO']
).dt.days

```

#Tratamento de Nulos


- Features
  - `VALOR_A_PAGAR`: Será imputado a mediana;
  - `RENDA_MES_ANTERIOR`: Será imputado a mediana;
  - `NO_FUNCIONARIOS`: Será imputado a mediana;
  - `SEGMENTO_INDUSTRIAL`: Será alterado para: "Desconhecido";
  - `DOMINIO_EMAIL`: Será alterado para: "Desconhecido";
  - `PORTE`: Será alterado para: "Desconhecido".

- Motivações
  - Para as features numéricas foi escolhido preencher com a mediana, pois é robusta a outliers, além de ser simples a sua imputação;
  - Para as features categóricas foi escolhido preencher com "Desconhecido", pois evita descartar linhas e perder os dados, como também preservar o fato de que a informação era ausente e fazer com que o modelo aprenda se essa ausência tem valor preditivo.


```python
#@title Imputando a mediana nas features "VALOR_A_PAGAR", "RENDA_MES_ANTERIOR" e "NO_FUNCIONARIOS"
features = ['VALOR_A_PAGAR', 'RENDA_MES_ANTERIOR', 'NO_FUNCIONARIOS']

for i in features:
  mediana_valor = df_cad_pag_info[i].median()
  df_cad_pag_info[i] = df_cad_pag_info[i].fillna(mediana_valor)
df_cad_pag_info.isnull().sum()
```

```python
#@title Imputando "Desconhecido" nas features categóricas com dados faltantes
cat_desconhecido = ['SEGMENTO_INDUSTRIAL', 'DOMINIO_EMAIL', 'PORTE']
df_cad_pag_info[cat_desconhecido] = df_cad_pag_info[cat_desconhecido].fillna('Desconhecido')
```

```python
#@title Lista para tratamento
outliers = ['VALOR_A_PAGAR', 'RENDA_MES_ANTERIOR', 'NO_FUNCIONARIOS']
```

```python
# Método: IQR capping

# Calcula o IQR, define os limites inferior e superior.
# Substitui os valores acima e abaixo desses limites pelos valores dos proprios limites.
# Cria três novas colunas com esses valores.

for col in outliers:
    Q1 = df_cad_pag_info[col].quantile(0.25)
    Q3 = df_cad_pag_info[col].quantile(0.75)
    IQR = Q3 - Q1
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR


    df_cad_pag_info[f'{col}_iqr_cap'] = df_cad_pag_info[col].clip(lower=lim_inf, upper=lim_sup)
```

```python
#@title Comparação visual direta entre as distribuições originais e as versões tratadas pelo IQR capping
caps    = [f'{v}_iqr_cap'         for v in outliers]

fig, axes = plt.subplots(nrows=len(outliers), ncols=4, figsize=(20, 5*len(outliers)))

for i, var in enumerate(outliers):
    orig = df_cad_pag_info[var]
    cap  = df_cad_pag_info[f'{var}_iqr_cap']

    # Histograma original
    sns.histplot(orig.dropna(), bins=30, kde=True, ax=axes[i,0], color='skyblue')
    axes[i,0].set_title(f'Hist Original: {var}')
    axes[i,0].grid(True, linestyle='--', alpha=0.3)

    # Histograma IQR‐capped
    sns.histplot(cap.dropna(), bins=30, kde=True, ax=axes[i,1], color='green')
    axes[i,1].set_title(f'Hist IQR Winsorized: {var}')
    axes[i,1].grid(True, linestyle='--', alpha=0.3)

    # Boxplot original
    sns.boxplot(x=orig.dropna(), ax=axes[i,2], color='lightcoral')
    axes[i,2].set_title(f'Box Original: {var}')
    axes[i,2].grid(True, linestyle='--', alpha=0.3)

    # Boxplot IQR‐capped
    sns.boxplot(x=cap.dropna(), ax=axes[i,3], color='mediumpurple')
    axes[i,3].set_title(f'Box IQR Winsorized: {var}')
    axes[i,3].grid(True, linestyle='--', alpha=0.3)


    for j in range(4):
        axes[i,j].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

  - Para as features categóricas foi escolhido preencher com "Desconhecido", pois evita descartar linhas e perder os dados, como também preservar o fato de que a informação era ausente e fazer com que o modelo aprenda se essa ausência tem valor preditivo.

# Modelo probabilístico

```python
#@title Carreagmento de bibliotecas

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#@title Leitura de bases

df_train = pd.read_csv('base_treino.csv')
df_test = pd.read_csv('base_teste.csv')
```

```python
#@title Separação de features e target
y_train = df_train['INADIMPLENTE'] # target
X_train = df_train.drop(['ID_CLIENTE', 'INADIMPLENTE','SAFRA_REF'], axis=1) # features treino
X_test = df_teste.drop(['ID_CLIENTE', 'SAFRA_REF'], axis=1) # features teste
```

```python
#@title Lista de features

num_cols = [
    'VALOR_A_PAGAR', 'TAXA', 'RENDA_MES_ANTERIOR',
    'NO_FUNCIONARIOS', 'dias_desde_emissao',
    'idade_cadastro_dias',
    'VALOR_A_PAGAR_iqr_cap', 'RENDA_MES_ANTERIOR_iqr_cap', 'NO_FUNCIONARIOS_iqr_cap'
]
cat_cols = [
    'FLAG_PF', 'SEGMENTO_INDUSTRIAL', 'DOMINIO_EMAIL',
    'PORTE', 'ESTADO'
]
```

```python
#@title Pré-processador que escalona features numéricas e executa one-hot encode nas features categóricas
preproc = ColumnTransformer([
    ('num', StandardScaler(), num_cols),
    ('cat', OneHotEncoder(handle_unknown='ignore', sparse_output=False), cat_cols)
])
```

```python
# Validação interna: split treino/validação
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train, y_train, test_size=0.2, stratify=y_train, random_state=42
)
```

```python
# Modelos a comparar
models = {
    'LogisticRegression': LogisticRegression(max_iter=1000),
    'RandomForest':       RandomForestClassifier(n_estimators=100, random_state=42),
    'GradientBoosting':   GradientBoostingClassifier(random_state=42),
    'AdaBoost':           AdaBoostClassifier(n_estimators=100, random_state=42),
    'KNeighbors':         KNeighborsClassifier(n_neighbors=5)
}
```

```python
# Avalia cada modelo por AUC
best_name, best_score, best_pipe = None, 0, None
for name, clf in models.items():
    pipe = Pipeline([('preproc', preproc), ('clf', clf)])
    pipe.fit(X_tr, y_tr)
    y_prob = pipe.predict_proba(X_val)[:,1]
    auc = roc_auc_score(y_val, y_prob)
    print(f"{name:17s} AUC = {auc:.4f}")
    if auc > best_score:
        best_score, best_name, best_pipe = auc, name, pipe
print(f"\nMelhor modelo: {best_name} com AUC = {best_score:.4f}")
```

```python
# Retreina o melhor no treino completo
best_pipe.fit(X_train, y_train)

# Gera probabilidades no teste “real”
probs = best_pipe.predict_proba(X_test)[:,1]
df_test['PROBABILIDADE_INADIMPLENCIA'] = probs
```

```python
#@title Matriz de confusão
y_pred = best_pipe.predict(X_val)
cm = confusion_matrix(y_val, y_pred)
tn, fp, fn, tp = cm.ravel()

print("Matriz de Confusão:")
print(pd.DataFrame(cm,
                   index=["real_0 (adimplente)","real_1 (inadimplente)"],
                   columns=["pred_0","pred_1"]))
print(f"\nTrue Positives  (inadimplentes corretamente previstos): {tp}")
print(f"False Negatives (inadimplentes perdidos):              {fn}")
print(f"False Positives (adimplentes classificados como 1):    {fp}")
print(f"True Negatives  (adimplentes corretos):                {tn}")
```

```python
#@title Relatório de métricas de classificação
# precision: previu a classe e quanto acertou
# recall: de todos os casos dreais quantos identificou corretamente
# f1: média harmônica entre precision e recal
# support: número de amostras
print(classification_report(y_val, y_pred, digits=4))
```

- Teve o maior AUC;
- Teve um alto recall;
- Gerou poucos FP;
- Em geral, foi o modelo que melhor aprendeu com o padrão de inadimplência do dataset.
